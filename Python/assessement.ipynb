```python
# Import the necessary Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the dataset
df = pd.read_csv("/content/Retail dataset of a global superstore F7.csv")
df
```

```python
#Show how many columns and rows
df.shape
```

```python
# Show descriptive statistics
df.describe()
```

```python
# Show dataset columns
df.columns
```

```python
# Print a concise summary of a DataFrame.
df.info()
```

```python
# Delete a column
df = df.drop(columns = "Row ID")
```

```python
# Print missing values
df.isnull().sum()
```

```python
# Format date to datetime type
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Extract month name from date
df['Month Name'] = df['Order Date'].dt.month_name()
# # Extract month number from date
df['Month Number'] = df['Order Date'].dt.month
```

```python
# Insert Quarter column
df['Quarter'] = df['Order Date'].dt.quarter
# Insert Year Column
df['year'] = df['Order Date'].dt.year
```

```python
# Fill missing values
df["Country"].fillna("United State", inplace = True)
```

```python
# Insert a new Column name 'Total Sales'
df['Total Sales'] = df['Unit Sale'] * df['Quantity']
```

```python
# Insert Total profit column
df['Total Profit'] = (df['Unit Sale'] - df['Unit Cost']) * df['Quantity']
```

```python
# print columns with the new created one
df.columns
```

```python
# Print the top 12 items
df.head(12)
```

```python
# Print data types
print(df.dtypes)
```

```python
# Formating data types
df['Country'] = df['Country'].astype('string')
df['Ship Mode'] = df['Ship Mode'].astype('string')
df['Customer ID'] = df['Customer ID'].astype('string')
df['Customer Name'] = df['Customer Name'].astype('string')
df['Segment'] = df['Segment'].astype('string')
df['City'] = df['State'].astype('string')
df['Region'] = df['Region'].astype('string')
df['Product ID'] = df['Product ID'].astype('string')
df['Category'] = df['Category'].astype('string')
df['Month Name'] = df['Country'].astype('string')
df['State'] = df['State'].astype('string')
df['Order ID'] = df['Order ID'].astype('string')
```

```python
df.info()
```

```python
# Rearrange the columns items
df = df[['Order ID', 'Order Date', 'year', 'Quarter', 'Month Name', 'Month Number', 'Ship Mode', 'Customer ID', 'Customer Name',
       'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region',
       'Product ID', 'Category', 'Unit Sale', 'Quantity', 'Unit Cost', 'Total Sales', 'Total Profit',
       'Month Name', 'Month Number']]
```

```python
# Print the rearranged column
df.columns
```

```python
#Check for duplicates rows
duplicates_row = df[df.duplicated()]
duplicates_row
```

```python
#1.	Calculate the total sales from the dataset.
totalSales = sum(df['Unit Sale'] * df['Quantity'])
totalSales = round(totalSales, 2)
print(totalSales)
```

```python
# 2.	Determine the average quantity sold for each product category.
average_quantity_sold = df.groupby('Category')['Quantity'].mean()
i = round(average_quantity_sold, 2)
print(i)
```

```python
# 3. Identify the most frequently used ship mode.
most_frequent_ship_mode = df['Ship Mode'].value_counts().idxmax()
print(f"The most frequently used ship mode is: {most_frequent_ship_mode}")
```

```python
# 4.	Find the top 5 customers based on their total sales.
customer_sales = df.groupby('Customer Name')['Total Sales'].sum()
top_customers = customer_sales.sort_values(ascending=False)
top_5_customers = top_customers.head(5)
print(top_5_customers)
```

```python
# 5.	Count the number of orders for each region
order_by_region = df.groupby('Region')['Order ID'].count()
print(order_by_region)
```

```python
# 6.	Calculate the average sales per quantity for the entire dataset.
total_sales = df['Total Sales'].sum()
total_quantity = df['Quantity'].sum()
average_sales_per_quantity = total_sales / total_quantity
print(f"Average Sales Per Quantity: {average_sales_per_quantity:.2f}")
```

```python
# 7.	Identify the order with the highest profit.

# Find the row with the maximum profit
max_profit_row = df.loc[df['Total Profit'].idxmax()]

# Extract the OrderID and Profit for this order
highest_profit_order_id = max_profit_row['Order ID']
highest_profit_amount = max_profit_row['Total Profit']

print(f"The order with the highest profit is OrderID: {highest_profit_order_id} with a profit of: £{highest_profit_amount:.2f}")
```

```python
# 8.	Count the number of orders that were shipped to the state of California.

# Filter df where the State is 'California'
orders_to_california = df[df['State'] == 'California']

# Count the number of orders
num_orders_to_california = len(orders_to_california)

print(f"Number of orders shipped to California: {num_orders_to_california}")
```

```python
#9.	Create a bar chart to visualize the total sales for each product category.
sales_by_category = df.groupby('Category')['Total Sales'].sum()

# Plotting
sales_by_category.plot(kind='bar', figsize=(8, 4), color='blue')

# Adding title and labels
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate category labels to avoid overlap

# Display the plot
plt.show()
```

```python
#10.	Generate a scatter plot to explore the relationship between sales and profit for each order.
plt.figure(figsize=(8, 4))  # Set the figure size
plt.scatter(df['Total Sales'], df['Total Profit'], color='blue')  # Plot sales vs. profit

# Adding title and labels
plt.title('Relationship between Sales and Profit')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')

# Show plot
plt.show()
```

```python
# 11.	Design a box plot to compare the distribution of quantity sold across different regions.
plt.figure(figsize=(8, 4))  # Set the figure size
sns.boxplot(x='Region', y='Quantity', data=df)

# Adding title and labels (optional)
plt.title('Distribution of Quantity Sold Across Different Regions')
plt.xlabel('Region')
plt.ylabel('Quantity')

# Show plot
plt.show()
```
